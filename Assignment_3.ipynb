```
import numpy as np
import matplotlib.pyplot as plt
```

```
overs = np.arange(1,21,3)
team1 = np.random.randint(0,150,7)
team1 = np.sort(team1)
team2 = np.random.randint(0,150,7)
team2 = np.sort(team2)
plt.plot(team1, marker = '*', c = 'r',label = 'Team 1')
plt.plot(team2, marker = '*', c = 'g',label = 'Team 2')
plt.scatter(team1,team2)
plt.xlabel('overs')
plt.ylabel('Runs')
plt.title('Analysis of cricket match')
plt.legend()
plt.show()
```
